```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

filepath = 'dataset'
dataset_x = pd.read_csv(filepath + '/dengue_features_train.csv')
dataset_y = pd.read_csv(filepath + '/dengue_labels_train.csv')

# pd.DataFrame(dataset_x)
# pd.DataFrame(dataset_y)
```

```python
x = dataset_x.iloc[:, :].values
y = dataset_y.iloc[:, -1].values
pd.DataFrame(x)
```

```python
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values='NaN', strategy='mean', axis=0)
imputer = imputer.fit(x[:, 4:])
x[:, 4:] = imputer.transform(x[:, 4:])
pd.DataFrame(x)
```

### Encoding categorical data

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelEncoder_x = LabelEncoder()
x[:, 0] = labelEncoder_x.fit_transform(x[:, 0])
oneHotEncoder = OneHotEncoder(categorical_features = [0])
x[:, 0] = oneHotEncoder.fit_transform([x[:, 0]]).toarray()
pd.DataFrame(x)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x[:, 4:] = sc_x.fit_transform(x[:, 4:])
pd.DataFrame(x)
```

### Fitting Polynomial Regression to the dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4) # try changing the degree
x_poly = poly_reg.fit_transform(x[:, 4:])
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_poly, y)
```

### Predicting a new result with Linear Regression
